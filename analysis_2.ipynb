-> Serve performance Comparison between losers & winners 


-> Parameters considered- Aces,Double Faults,First Serve Points Won, Second Serve Points Won


```python
import pandas as pd
matches_copy=pd.read_csv("/Users/ssweeti/Desktop/Tennis-project_EDA/tennis_atp-master/matches_cleaned.csv")
```

```python
import matplotlib.pyplot as plt
import numpy as np

## serve performance comparison between winners and losers

serve_compare = matches_copy[[
    "w_ace","l_ace",
    "w_df","l_df",
    "w_1stIn","l_1stIn",
    "w_1stWon","l_1stWon"
]].mean()


## avergae serve performance 

serve_compare = pd.DataFrame({
    "Winner Avg":[
        serve_compare["w_ace"],
        serve_compare["w_df"],
        serve_compare["w_1stIn"],
        serve_compare["w_1stWon"]
    ],
    "Loser Avg":[
        serve_compare["l_ace"],
        serve_compare["l_df"],
        serve_compare["l_1stIn"],
        serve_compare["l_1stWon"]
    ]
},
index=["Aces","Double Faults","1st Serve In","1st Serve Won"]
)

#Figure
fig, ax = plt.subplots(figsize=(8,6))  

# Background Styling
fig.patch.set_facecolor("#061A2B") 
ax.set_facecolor("#061A2B")

#  Color Palette
winner_color = "#00C6FF"   
loser_color = "#F9E547"    

# Plot bar chart
serve_compare.plot(
    kind="bar",
    ax=ax,
    color=[winner_color, loser_color],
    width=0.6,
    edgecolor="white",
    linewidth=0.8
)

# title styling
ax.set_title(
    "Serve Performance: Winners vs Losers",
    fontsize=15,
    fontweight="bold",
    color="white",
    pad=15
)

# Axis Labels
ax.set_ylabel("Average Value", color="white", fontsize=11)
ax.set_xlabel("Serve Metrics", color="white", fontsize=11)

# Tick Styling
ax.tick_params(colors="white")
plt.xticks(rotation=25, ha="right")

# Grid Styling
ax.grid(axis="y", linestyle="--", alpha=0.25, color="white")
ax.set_axisbelow(True)

# Borders removal
for spine in ax.spines.values():
    spine.set_visible(False)

# Value Labels
for container in ax.containers:
    ax.bar_label(
        container,
        fmt="%.1f",
        fontsize=10,
        color="white",
        padding=3
    )

#Legend Styling 
legend = ax.legend(
    ["Winner Avg", "Loser Avg"],
    fontsize=10,
    facecolor="#061A2B",
    edgecolor="none"
)

for text in legend.get_texts():
    text.set_color("white")

plt.tight_layout()
plt.show()
```
