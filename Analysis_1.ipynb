->> Analysis of top 10 players by numbers of match wins

Import required libraries & loading the dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
matches_copy=pd.read_csv("/Users/ssweeti/Desktop/Tennis-project_EDA/tennis_atp-master/matches_cleaned.csv")
```

Calculate total wins by Each player

```python
player_wins = (
    matches_copy.groupby("winner_name")
      .size()
      .reset_index(name="wins")
      .sort_values("wins", ascending=False)
)
```

Choose top 10 players

```python
top10 = player_wins.head(10)
```

Creating the Bar chart for top 10 players with number of wins

```python
plt.figure(figsize=(12,6))

## color gradient 
colors = plt.cm.viridis(np.linspace(0, 1, len(top10)))

bars = plt.bar(top10["winner_name"], top10["wins"], color=colors)

# x-axis labels rotate
plt.xticks(rotation=45, ha="right")

plt.title("Top 10 Players by Match Wins (2020–2024)", fontsize=16, fontweight="bold")
plt.xlabel("Players")
plt.ylabel("Number of Wins")

# Grid
plt.grid(axis="y", linestyle="--", alpha=0.5)

# Add value labels
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2,
             height + 2,
             int(height),
             ha="center")

plt.tight_layout()
plt.show()
```
