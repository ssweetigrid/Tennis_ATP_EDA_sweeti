#  ATP Tennis Match Data Analysis (2020–2024)


## Introduction:
Tennis has evolved into a highly competitive and data-driven sport where match outcomes depend on multiple performance factors such as serving efficiency, player rankings, and match conditions. This project focuses on analyzing ATP Men’s Singles match data from 2020 to 2024 to uncover patterns and insights that define modern professional tennis.

The dataset includes match-level information from ATP tournaments across different surfaces, tournament categories, and competition rounds. It also contains player rankings and performance statistics such as aces, double faults, service points, break-point performance, and match duration.

Through Exploratory Data Analysis (EDA), this project aims to:

- Understand key performance metrics influencing match results  
- Identify trends in match competitiveness and duration  
- Analyze the impact of player rankings and seeding  
- Compare performance across different playing surfaces  
- Discover patterns that contribute to player success  

The insights generated from this analysis can help coaches, analysts, and tennis enthusiasts better understand performance strategies in modern ATP tennis.

```python
## match duration analysis (missing data filled with median)

matches_copy['minutes'].fillna(matches_copy['minutes'].median(), inplace=True)  
```

```python
## match statistsics columns missing data filled with 0 

stat_cols = [col for col in matches_copy.columns if col.startswith('w_') or col.startswith('l_')]

matches_copy[stat_cols] = matches_copy[stat_cols].fillna(0)
```

```python
## player heights data handle

matches_copy['winner_ht'].fillna(matches_copy['winner_ht'].median(), inplace=True)
matches_copy['loser_ht'].fillna(matches_copy['loser_ht'].median(), inplace=True)
```

```python
## winnner & loser rank points

matches_copy['winner_rank_points']= matches_copy['winner_rank_points'].fillna(0)
matches_copy['loser_rank_points']= matches_copy['loser_rank_points'].fillna(0)
```

```python
matches_copy.isna().sum()
```

```python
matches_copy['surface'] = matches_copy['surface'].fillna(matches_copy['surface'].mode()[0])
```

```python
matches_copy['winner_age'] = matches_copy['winner_age'].fillna(matches_copy['winner_age'].median())
```

```python
matches_copy['loser_age'] = matches_copy['loser_age'].fillna(matches_copy['loser_age'].median())
```
